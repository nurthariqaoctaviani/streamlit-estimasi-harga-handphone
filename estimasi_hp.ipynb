# Visualisasi Data

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

```python
models = df.groupby('battery_power').count()[['ram']].sort_values(by='ram',ascending=True).reset_index()
models = models.rename(columns={'ram':'BatteryRAM'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['battery_power'], y=models['BatteryRAM'], color='royalblue')
plt.xticks(rotation=60)
```

# distribusi harga handphone

```python
plt.figure(figsize=(15,5))
sns.distplot(df['price_range'])
```

# seleksi fitur

```python
features = ['battery_power','fc','int_memory','n_cores','ram']
x = df[features]
y = df['price_range']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.33, random_state=110)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score, '%')
```

# membuat inputan model regresi linier

```python
input_data = np.array([[842,1,7,2,2549]])

prediction = lr.predict(input_data)
print('Estimasi kisaran harga Handphone :', prediction, 'Juta')
```
